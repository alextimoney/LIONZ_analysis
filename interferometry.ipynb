```python
import matplotlib.pyplot as plt
import tifffile
from PIL import Image
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
import matplotlib.transforms as mtransforms
import cmcrameri.cm as cm
```

```python
with Image.open('LIONZ_zshot3_raw/interferometry/shot/z4290-simx-shot.TIF') as img:
    z4290 = np.array(img)
```

```python
fig, axs, = plt.subplots()
plt.imshow(z4290, cmap='gray')
axs.set_axis_off()
plt.savefig('interferometry_images/frame.jpg', bbox_inches = 'tight')
```

```python
y_ext, x_ext = z4290.shape[:2]

mid_y = y_ext // 2
mid_x = x_ext // 2

# 3. Slice the array into four separate images
top_left     = z4290[0:mid_y, 0:mid_x]
top_right    = z4290[0:mid_y, mid_x:x_ext]
bottom_left  = z4290[mid_y:y_ext, 0:mid_x]
bottom_right = z4290[mid_y:y_ext, mid_x:x_ext]

plt.imshow(top_left, cmap = 'gray')
plt.title('top left')
plt.savefig('top_left', bbox_inches = 'tight')
plt.clf()

plt.imshow(top_right, cmap = 'gray')
plt.title('top right')
plt.savefig('top_right',bbox_inches = 'tight')
plt.clf()

plt.imshow(bottom_left, cmap = 'gray')
plt.title('bottom left')
plt.savefig('bottom_left',bbox_inches = 'tight')
plt.clf()

plt.imshow(bottom_right, cmap = 'gray')
plt.title('bottom right')
plt.savefig('bottom_right',bbox_inches = 'tight')
plt.clf()
```
